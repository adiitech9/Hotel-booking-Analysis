## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# lets check missing values
miss_value[:10]
```

```python
# % of null value in each column
percentage_company_null_value = miss_value[0] / unique_rows * 100
percentage_company_null_value
```

```python
# % values of agent
percentage_agent_null_value = miss_value[1] / unique_rows * 100
percentage_agent_null_value
```

```python
# % values of country
percentage_country_null_value = miss_value[2] / unique_rows * 100
percentage_country_null_value
```

```python
# we have less null values in children & country column, so we replace null from 0 as country column
hotel_booking['children'].fillna(0, inplace=True)
hotel_booking['country'].fillna(0, inplace=True)
print(hotel_booking[['children', 'country']].isnull().sum())
```

```python
# Check if 'company' and 'agent' columns exist before dropping
if 'company' in hotel_booking.columns and 'agent' in hotel_booking.columns:
    hotel_booking = hotel_booking.drop(['company', 'agent'], axis=1)
else:
    print("Columns 'company' and/or 'agent' are not present in the DataFrame.")
```

```python
hotel_booking = hotel_booking.dropna(axis=0)
```

```python
hotel_booking.shape
```

```python
# Visualizing the missing values
sns.heatmap(hotel_booking.isnull(), cbar=False)
```

```python
hotel_booking.isnull().sum()
```

```python
# total stay in nights
hotel_booking['total_stay_in_nights'] = hotel_booking['stays_in_weekend_nights'] + hotel_booking['stays_in_week_nights']
hotel_booking['total_stay_in_nights']
```

```python
# created column for revenue using total_stay * adr
hotel_booking['revenue'] = hotel_booking['total_stay_in_nights'] * hotel_booking['adr']
hotel_booking['revenue']
```

```python
# total guest coming for each booking
hotel_booking['total_guest'] = hotel_booking['adults'] + hotel_booking['children'] + hotel_booking['babies']
hotel_booking['total_guest'].sum()
```

```python
# from column is_cancelled , we will replace the value from(0,1) to not_cancelled,is cancelled
hotel_booking['is_cancelled']= hotel_booking['is_canceled'].replace([0,1],['not_cancelled','is_cancelled'])
hotel_booking['is_cancelled']
```

```python
# same repeated guest column
hotel_booking['is_repeated_guest'] = hotel_booking['is_repeated_guest'].replace([0,1],['not_repeated','repeated'])
hotel_booking['is_repeated_guest']
```

```python
# revenue hotel wise
total_revenue = hotel_booking.groupby('hotel')['revenue'].sum()
total_revenue
```

### What all manipulations have you done and insights you found?

1. Derived Columns:

Total Guests: This column helps evaluate the total number of guests, calculated as the sum of adults, children, and babies.

Revenue: Revenue is derived by multiplying the ADR (Average Daily Rate) by the Total Guests. This helps analyze the profit and growth of each hotel.

0 → "Not Canceled", 1 → "Canceled" (for cancellation status)

0 → "Not Repeated", 1 → "Repeated" (for repeated guest status)
This improves readability and understanding during visualization.



4. Data Type Conversion:

Agent & Children: These columns contained float values, which didn't make sense as they represent guest counts and agent IDs. We converted them from float to integer.


5. Handling Missing and Duplicate Data:

Null Values: We checked for missing values across all columns:

Columns with a high number of null values were dropped (e.g., "company").

Columns with minimal null values were filled using appropriate methods like .fillna().

```python
# Chart - 2 visualization code
def get_count_from_column_bar(df, column_label):
  hotel_booking_grpd = hotel_booking[column_label].value_counts()
  hotel_booking = pd.DataFrame({'index':hotel_booking_grpd.index, 'count':hotel_booking_grpd.values})
  return hotel_booking

  def plot_bar_chart_from_column(hotel_booking, column_label, t1):
    hotel_booking_grpd = get_count_from_column(hotel_booking, column_label)
    fig, ax = plt.subplots(figsize=(25,8))
    c= ['g','r','b','c','y']
    ax.bar(hotel_booking_grpd['index'],hotel_booking_grpd['count'], width= 0.6, align= 'edge', edgecolor= 'black', linewidth=6 , color = c, linestyle= ":",alpha= 0.6)
    plt.title(t1, bbox={'facecolor':'0.20','pad':6})
    plt.legend()
    plt.ylable('count')
    plt.xticks(rotation=20)
    plt.xlabel(column_label)
    plt.show()
```

```python
# chart - 2 visulization code
def get_count_from_column(hotel_booking, column_label):
  hotel_booking_grpd = hotel_booking[column_label].value_counts()
  hotel_booking = pd.DataFrame({'index':hotel_booking_grpd.index, 'count':hotel_booking_grpd.values})
  return hotel_booking

def plot_pie_chart_from_column(hotel_booking, column_label, t1, exp):
    hotel_booking_grpd = get_count_from_column(hotel_booking, column_label)
    fig, ax = plt.subplots(figsize=(20,8))
    ax.pie(hotel_booking_grpd.loc[:, 'count'], labels=hotel_booking_grpd.loc[:, 'index'], autopct='%1.2f%%', startangle=90,shadow=True,labeldistance = 1, explode = exp)
    plt.title(t1, bbox={'facecolor':'0.8','pad':6})
    ax.axis('equal')
    plt.legend()
    plt.show()

exp = [0.08,0.08]
plot_pie_chart_from_column(hotel_booking, 'hotel', 'Booking Percentage Of Hotel By Name', exp)
```
